[문제 1]

사용자로부터 두 개의 숫자를 입력받아 나눗셈을 수행하는 프로그램을 작성하세요.

[예외처리]
- 입력 값이 숫자가 아닐 때 발생할 수 있는 ValueError
- 나누는 수가 0일 때 발생할 수 있는 ZeroDivisionError
- 위 2가지 예외처리 후 사용자에게 이해하기 쉬운 메시지 작성 후 출력되도록 코드를 작성

````python
def divide_num():

  try:
      number_first = int(input("첫 번째 숫자를 입력하세요.: "))
      number_second = (int(input("두 번째 숫자를 입력하세요.: ")))

      result = number_first / number_second #자료형이 float으로 나오는 듯 함

  except ValueError:
    print("입력된 값이 숫자가 아닙니다.")
  except ZeroDivisionError:
    print("0으로 나눌 수 없습니다.")

  else:
    print(f"결과는 {result} 입니다.")

  finally:
    print("실행을 종료합니다.")

divide_num()
````

````python
def read_file():
  try:
    file_name = input("파일 이름을 입력하세요.: ")

    with open(file_name, 'r'):
      print("파일이 있는지 탐색합니다.")

  except FileNotFoundError:
    print("파일을 찾을 수 없습니다.")

  else:
    print("파일을 읽었습니다.")

  finally:
    print("실행을 종료합니다.")

read_file()
````

````python
class UserException(Exception):
  pass

class NegativeNumberError(UserException):
  pass

def entry_number():

  num = int(input("숫자를 입력해주세요.: "))

  try:
    if num < 0:
      raise NegativeNumberError("음수를 입력하셨습니다.")
  except NegativeNumberError as e:
      print("양수를 입력해주세요.")

  else:
    print("양수를 입력하셨습니다.")

  finally:
    print("프로그램을 종료합니다.")

entry_number()
````

[문제 7]

함수 실행 시간을 측정하는 데코레이터 timer_decorator를 작성하세요.

임의의 함수가 실행되는 시간을 측정하여 "함수 실행 시간: x.xx초" 형태로 출력하세요.

````python
import time
import functools

def timer_decorator(func):
    @functools.wraps(func)
    def wrapper(*args, **kwargs):
        start_time = time.perf_counter()

        result = func(*args, **kwargs)

        end_time = time.perf_counter()

        execution_time = end_time - start_time
        print(f"함수 실행 시간: {execution_time:.2f}초")

        return result
    return wrapper

@timer_decorator
def heavy_computation():
    print("작업을 시작합니다.")
    time.sleep(3)
    print("작업 완료")

heavy_computation()
````

````python
# mymath.py
import math

def triangle_area(base, height):
    return 0.5 * base * height

def circle_area(radius):
    return math.pi * (radius ** 2)

def cuboid_surface_area(width, length, height):
    return 2 * (width * length + length * height + height * width)
````

````python
# main.py
import mymath

tri_area = mymath.triangle_area(10, 5)
print(f"삼각형의 넓이: {tri_area}")

circ_area = mymath.circle_area(7)
print(f"원의 넓이: {circ_area:.2f}")

cub_surface = mymath.cuboid_surface_area(3, 4, 5)
print(f"직육면체의 겉넓이: {cub_surface}")
````
