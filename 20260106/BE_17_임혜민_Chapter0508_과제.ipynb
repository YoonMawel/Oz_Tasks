__[문제 1: 첫 번째 사냥]__

마법사는 몬스터를 사냥하기 위해 마법을 사용합니다.

마법을 사용하려면 지능이 50 이상이어야 합니다.

마법사의 지능이 50 이상인지 확인하세요.

```python
def check_magic():
  #마법사의 스테이터스 생성 및 읽어오기
  magician = {"strength":10, "dexterity":20, "intelligence":80, "hp":50}
  print(f'마법사의 지능은 {magician.get('intelligence')} 입니다.')

  #마법사의 지능 수치를 조건문으로 판단
  if magician.get('intelligence') >= 50:
    print('지능이 50 이상입니다. 마법을 사용할 수 있습니다.')
  else:
    print('지능이 50 미만입니다. 마법을 사용할 수 없습니다.')

check_magic()
```

__[문제 2: 두 번째 사냥]__

전사는 강력한 공격을 하기 위해 자신의 힘을 사용합니다.

전사의 힘이 60 이상이고 체력이 80 이상인지 확인하세요.

```python
def check_war():
  #전사의 스테이터스 생성 및 읽어오기
  #전사: 힘(80), 민첩(30), 지능(20), 체력(100)
  warrior = {"strength":80, "dexterity":30, "intelligence":20, "hp":100}
  print(f'전사의 힘은 {warrior.get('strength')} 입니다.')
  print(f'전사의 체력은 {warrior.get('hp')} 입니다.')

  #전사의 힘 수치를 판단, 힘 >= 60, 체력 >= 80
  if warrior.get('strength') >= 60 and warrior.get('hp') >= 80:
    print('전사는 강력한 공격을 할 수 있습니다.')
  elif warrior.get('strength') < 60:
    print('전사의 힘이 부족합니다.')
  elif warrior.get('hp') < 80:
    print('전사의 체력이 부족합니다.')
  else:
    print(f'전사는 강력한 공격을 할 수 없습니다.')

check_war()
```

__[문제 3: 세 번째 사냥]__

도적은 민첩함을 이용해 몬스터를 회피하고 공격합니다.

도적의 민첩이 70 이상이거나 힘이 50 이상인지 확인하세요.

```python
def check_bandit():
  #도적의 스테이터스 생성 및 읽기
  #도적: 힘(40), 민첩(80), 지능(30), 체력(60)
  bandit = {"strength": 40, "dexterity":80, "intelligence":30, "hp": 60}
  print(f'도적의 민첩은 {bandit.get('dexterity')}, 힘은 {bandit.get('strength')}입니다.')

  #도적의 회피 조건 판단
  if bandit.get('dexterity') >= 70 and bandit.get('strength') >= 50:
    print('도적은 민첩하게 몬스터의 공격을 피했습니다.')
  elif bandit.get('dexterity') < 70:
    print(f'몬스터의 공격을 피할 수 없습니다. 민첩함이 부족합니다. / 현재 민첩: {bandit.get('dexterity')}')
  elif bandit.get('strength') < 50:
    print(f'몬스터의 공격을 피할 수 없습니다. 힘이 부족합니다. / 현재 힘: {bandit.get('strength')}')
  else:
    print('몬스터의 공격을 피할 수 없습니다.')

check_bandit()
```

__[문제 4: 네 번째 사냥]__

힐러는 파티의 체력을 회복시키기 위해 마법을 사용합니다.

힐러의 지능이 60 이상이고, 체력이 50 이상인지 확인하세요.

```python
def check_healer():
  #힐러의 스테이터스 생성 및 읽기
  #힐러: 힘(20), 민첩(40), 지능(70), 체력(70)
  healer = {"strength": 20, "dexterity": 40, "intelligence": 70, "hp": 70}
  print(f'힐러의 지능은 {healer.get('intelligence')}이며, 체력은 {healer.get('hp')}입니다.')

  #힐러의 회복 마법 조건 판단
  if healer.get('intelligence') >= 60 and healer.get('hp') >= 50:
    print('힐러는 팀원을 회복시킬 수 있습니다.')
  elif healer.get('intelligence') < 60:
    print(f'힐러의 지능이 부족합니다. 회복을 할 수 없습니다. / 현재 지능: {healer.get('intelligence')}')
  elif healer.get('hp') < 50:
    print(f'힐러의 체력이 부족합니다. 회복을 할 수 없습니다. / 현재 체력: {healer.get('hp')}')
  else:
    print('힐러는 팀원을 회복시킬 수 없습니다.')

check_healer()
```

__[문제 5: 레벨업]__

모든 캐릭터가 사냥을 성공적으로 마쳤습니다. 이제 경험치를 얻어 레벨업을 해야 합니다.


각 캐릭터의 경험치가 100 이상인지 print()를 이용해 확인해주세요.

<br>

경험치는 아래와 같습니다.

마법사: 120

전사: 150

도적: 110

힐러: 90



```python
def level_up():
  #팀원 별 경험치 생성 및 읽기
  exp = {"magician": 120, "warrior": 150, "bandit": 110, "healer": 90}
  #print(f'마법사의 경험치는 {exp.get('magician')}입니다.')
  #print(f'전사의 경험치는 {exp.get('warrior')}입니다.')
  #print(f'도적의 경험치는 {exp.get('bandit')}입니다.')
  #print(f'힐러의 경험치는 {exp.get('healer')}입니다.')

  #조회하고 싶은 파티원을 직접 입력
  name = input('어떤 팀원의 경험치가 궁금한가요?: ')

  #사용자가 입력한 파티원의 이름을 파싱? 변환
  if name == '마법사':
    name = 'magician'
  elif name == '전사':
    name = 'warrior'
  elif name == '도적':
    name = 'bandit'
  elif name == '힐러':
    name = 'healer'
  else:
    print('해당 팀원은 존재하지 않습니다.')
    return #else로 판단 시 해당 단계에서 실행 종료

  #해당 파티원의 경험치를 출력, 레벨업 조건을 같이 판단 (이중if문)
  if name in exp.keys():
    print(f'{name}의 경험치는 {exp[name]}입니다.')
    if exp[name] >= 100:
      print(f'{name}은 레벨업을 할 수 있습니다.')
    else:
      print(f'{name}은 레벨업을 할 수 없습니다.')
  else:
    print('해당 팀원은 존재하지 않습니다. 2')

level_up()
```
