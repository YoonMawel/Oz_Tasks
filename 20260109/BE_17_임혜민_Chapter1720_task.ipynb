__[문제1]__

print_even 리스트 내에 있는 값 중 짝수만 화면에 출력되도록 print_score 함수를 생성 후 결과값을 출력해주세요

print_even ([1, 3, 2, 10, 12, 11, 15])

```python
print_even = [1, 3, 2, 10, 12, 11, 15]

def print_score(print_even):
  for i in print_even:
    if i % 2 == 0:
      print(i)

print_score(print_even)
```

__[문제7]__

오즈 (oz) 클래스에서 이름, 나이, 성별을 출력하는 who() 메소드를 추가하고 who()메소드를 이용해 아래 출력 결과를 만들어주세요


출력 결과 ) <br>
이름: 김코, 나이: 30, 코딩레벨: A

```python
class Oz:
  def who(self, name, age, level):
    self.name = name
    self.age = age
    self.level = level

    print(f"이름: {self.name}\n 나이: {self.age} \n 코딩레벨: {self.level}")

result = Oz()
result.who("김코", 30, "A")
```
